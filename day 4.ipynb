```python
def Is_Prime(n):
    cnt=0
    for i in range(1,n+1):
        if n%i==0:
            cnt+=1
    if cnt==2:
        print(n,'is prime number')
    else:
        print(n,'is not prime number')
start=int(input('enter start range:'))
end=int(input('enter end range:'))
prime_list=[i for i in range(start,end+1) if Is_Prime(i)]
```
